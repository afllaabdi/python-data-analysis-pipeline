```python
import numpy as np
course_data = ({
    "Data Mining": 85,
    "Machine Learning": 90,
    "Artificial Intelligence": 88,
    "Database Systems": 92,
    "Computer Networks": None,
    "Operating Systems": "",
    "Software Engineering": 80,
    "Web Programming": 75,
    "Cyber Security": None,
    "Statistics": 89
},)

print("Data Awal:")
print(course_data)
```

```python
courses = course_data[0]

cleaned_courses = {k: v for k, v in courses.items() 
                   if v not in [None, ""]}

print("\nAfter Hapus nilai kosong(None atau ''):")
print(cleaned_courses)
```

```python
jumlah= len(cleaned_courses)
print("\nJumlah mata kuliah yg tersisa:", jumlah)
```

```python
sorted_courses = dict(sorted(cleaned_courses.items(), 
                             key=lambda item: item[1]))
print("\nSetelah sorting nilai(ascending):")
print(sorted_courses)
```

```python
nilai_tertinggi = max(cleaned_courses, key=cleaned_courses.get)

print("\nMata kuliah dengan nilai tertinggi:", nilai_tertinggi)
print("Nilanya:", cleaned_courses[nilai_tertinggi])
```

```python
average_grade = sum(cleaned_courses.values()) / len(cleaned_courses)

print("\nRata-rata nilai:", average_grade)
```

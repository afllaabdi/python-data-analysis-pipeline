```python
import random
def proses_list(input_list):
    data = input_list.copy()
    print("List awal:", data)

    for i in range(10):
        if i % 2 == 0:
            data.append(random.randint(1, 100))
        else:
            data.append(random.uniform(1, 100))
    
    print("\nAfter menambahkan 10 angka acak:")
    print(data)

    data = [x for i, x in enumerate(data) if i % 2 != 0]

    print("\nSetelah menghapus elemen pada indeks genap:")
    print(data)

    data.reverse()

    print("\nAfter Reverse:")
    print(data)

    nilai_terbesar = max(data)
    data.remove(nilai_terbesar)

    print("\nSetelah menghapus nilai terbesar:")
    print(data)

    data.sort(reverse=True)

    print("\nSetelah sorting descending:")
    print(data)

    return data

my_list = [10, 20, 30, 40, 50]
hasil = proses_list(my_list)

print("\nHasil akhir:", hasil)
```

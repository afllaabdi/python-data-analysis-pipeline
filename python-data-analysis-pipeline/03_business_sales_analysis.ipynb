```python
import pandas as pd

data = {
    "Merek": ["Avoskin", "G2G", "Skintific", "Wardah", "Emina",
              "Sesuatu", "Azarin", "Lacoco", "ANDA", "Nyonya Gie"],
    "Cabang": ["Surabaya", "Jakarta Pusat", "Bandung", "Bandung", "Jakarta",
               "Bandung", "Surabaya", "Bandung", "Jakarta Pusat", "Surabaya"],
    "Harga (ribuan)": [150, 75, 120, 85, 75, 120, 50, 90, 55, 50],
    "Saham": [175, 160, 120, 300, 230, 200, 180, 100, 180, 200],
    "Status": ["Terjual", "Tersedia", "Tersedia", "Terjual", "Tersedia",
               "Tersedia", "Tersedia", "Terjual", "Tersedia", "Terjual"]
}

df = pd.DataFrame(data)

print(df)
```

```python
print(df.describe())

print("\nJumlah Harga:", df["Harga (ribuan)"].sum())
print("Jumlah Saham:", df["Saham"].sum())
```

```python
df["Pendapatan"] = df["Harga (ribuan)"] * df["Saham"]

top3 = df.sort_values(by="Pendapatan", ascending=False).head(3)

print("\n Top 3 merek pendapatan tertinggi:")
print(top3[["Merek", "Pendapatan"]])
```

```python
group_harga = df.groupby("Harga (ribuan)")["Merek"].apply(list) 

print("\nMerek berdasarkan harga:")
print(group_harga)
```

```python
Q1 = df["Harga (ribuan)"].quantile(0.25)
Q3 = df["Harga (ribuan)"].quantile(0.75)

def kategori(harga):
    if harga < Q1:
        return "Murah"
    elif harga > Q3:
        return "Mahal"
    else:
        return "Sedang"
    
df["Kategori"] = df["Harga (ribuan)"].apply(kategori)

print("\nData dengan kategori harga:")
print(df[["Merek", "Harga (ribuan)", "Kategori"]])
```
